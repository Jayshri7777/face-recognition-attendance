```python
import pickle
```

```python
with open('/kaggle/input/face-recognition-using-vgg/faces_data.pkl', 'rb') as f:
    face_data = pickle.load(f)
```

```python
print("Type of face_data:", type(face_data))
```

```python
face_array = np.array(face_data)
```

```python
flat_faces = face_array.reshape(len(face_array), -1)
```

```python
df = pd.DataFrame(flat_faces)
```
